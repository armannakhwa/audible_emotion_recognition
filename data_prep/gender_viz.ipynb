```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
filepath = "data/visualization_data.csv"

df = pd.read_csv(filepath)
df.head()
```

```python
                      
gender_group = df.groupby(["gender"])

gender_df = gender_group.mean()
gender_df
```

```python
gender_df.drop(inplace=True, columns=(["label", "Unnamed: 0"]))
gender_dict= gender_df.to_dict('index')
gender_dict
```

```python
new_df = pd.DataFrame.from_dict(gender_dict)
new_df
```

```python
new_df.columns = ['Masculine', 'Feminine']
new_df
```

```python
new_df.drop('0', axis=0, inplace=True)
new_df
```

```python
new_df.to_csv("data/gender_column.csv", encoding='utf-8', index=True)
```

```python
male = new_df.plot.line(y='Masculine', figsize=(10, 4), color='black', title="Masculine")
male.set_xlabel("Feature")
male.set_ylabel("Frequency")

fig = male.get_figure()
fig.savefig("static/img/male_line.png")
```

```python
female = new_df.plot.line(y='Feminine', figsize=(10, 4), color='magenta', title="Feminine")
female.set_xlabel("Feature")
female.set_ylabel("Frequency")
fig = female.get_figure()
fig.savefig("static/img/female_line.png")
```
