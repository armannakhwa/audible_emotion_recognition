```python
import pandas as pd
import numpy as np
import matplotlib as plt

```

```python
filepath = "data/visualization_data.csv"

df = pd.read_csv(filepath)
df.head()
```

```python

                       
emotion_group = df.groupby(["emotions"])

emotion_df = emotion_group.mean()
emotion_df
```

```python
emotion_df.to_csv("data/emotion_avg.csv", encoding='utf-8', index=True)
```

```python
emotion_df.drop(inplace=True, columns=(["gender", "label", "Unnamed: 0"]))
emo_dict= emotion_df.to_dict('index')
emo_dict
```

```python
new_df = pd.DataFrame.from_dict(emo_dict)
new_df
```

```python
new_df.drop('0', axis=0, inplace=True)
new_df
```

```python
new_df.to_csv("data/emo_column.csv", encoding='utf-8', index=True)
```

```python
angry = new_df.plot.line(y='angry', figsize=(10, 4), color='red', title="Angry")
angry.set_xlabel("Feature")
angry.set_ylabel("Frequency")
#plt.savefig("Viz_Images/fear_line.png")
fig = angry.get_figure()
fig.savefig("Viz_Images/angry_line.png")
```
